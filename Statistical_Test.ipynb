```
with open("/content/drive/My Drive/RNN/Stats_RNN/RNN_lateral/Debris_30/predictions_RNN_Lateral_Debris_30.txt", "rb") as fp:
  model_1 = pickle.load(fp)

with open("/content/drive/My Drive/RNN/Stats_RNN/RNN_lateral/Debris_50/predictions_RNN_Lateral_Debris_50.txt", "rb") as fp:
  model_2 = pickle.load(fp)
```

```
test_images = loadmat('/content/drive/My Drive/RNN/Test/test_images_with_debris_10.mat')
y_true = test_images["targets"]
```

```
bool_array = tf.cast(tf.equal(y_true, model_1), tf.float32).numpy()
bool_array1 = tf.cast(tf.equal(y_true, model_2), tf.float32).numpy()
```

```
def contingency_table(data1, data2):
  fc = cf = ff = cc =0
  for i in range(len(data1)):
    if data1[i]==0 and data2[i] == 1:
      fc+=1
    elif data2[i]==0 and data1[i]==1:
      cf+=1
    elif data1[i]==0 and data2[i]==0:
      ff+=1
    else:
      cc+=1
  contingency_table = np.array([[cc, fc],[cf, ff]])
  print(contingency_table)
  return contingency_table
```

```
c_mat = contingency_table(bool_array, bool_array1)
```

```
import statsmodels.stats.contingency_tables as ct
import statsmodels.formula.api as smf
stat = ct.mcnemar(c_mat, exact=False, correction=False)
```

```
print(stat)
```
